```python
import os 
import pandas as pd 
```

```python
df = pd.DataFrame(pd.read_csv("/Users/savannahposner/Desktop/Big_Data/Vine_table.CSV"))
df
```

```python
df_2 = df.loc[df['total_votes'] >= 20]

df_2["helpful/total"] = (
    df_2["helpful_votes"]/df_2["total_votes"]
)
df_2
```

```python
df_3 = df_2.loc[df_2['helpful/total'] >= .5]
df_3
```

```python
df_vine_yes = df_3.loc[df_3['vine']=='Y']
df_vine_yes
```

```python
df_vine_no = df_3.loc[df_3['vine']=='N']
df_vine_no
```

```python
df_vine_yes_count = len(df_vine_yes['review_id'])
# df_vine_yes_count
five_star_reviews = len(df_vine_yes.loc[df_vine_yes["star_rating"] == 5])
# five_star_reviews
five_star_percentage = (round(five_star_reviews/df_vine_yes_count,2)*100)
# five_star_percentage
print(f'Vine Yes Table \n'
      f'The total vote count: {df_vine_yes_count} \n'
      f'The five star review count: {five_star_reviews} \n' 
      f'The five star percentage: {five_star_percentage}% \n\n')

df_vine_no_count = len(df_vine_no['review_id'])
# df_vine_yes_count
five_star_reviews_ = len(df_vine_no.loc[df_vine_no["star_rating"] == 5])
# five_star_reviews
five_star_percentage_ = (round(five_star_reviews_/df_vine_no_count,2)*100)
# five_star_percentage
print(f'Vine No Table \n'
      f'The total vote count: {df_vine_no_count} \n'
      f'The five star review count: {five_star_reviews_} \n' 
      f'The five star percentage: {five_star_percentage_}%')
```
